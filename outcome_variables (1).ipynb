```python
# Define the columns to keep
columns_to_keep = [
    "person_id", 
    "standard_concept_name", 
    "condition_start_datetime", 
    "condition_end_datetime", 
    "condition_type_concept_name", 
    "visit_occurrence_concept_name"
]

# Filter the dataset
cancer_conditions_df = dataset_48335132_condition_df[columns_to_keep]

# Display the first few rows
cancer_conditions_df.head()

```

```python
# Step 1: Get the first cancer diagnosis per person
y_cases = (
    cancer_conditions_df.groupby("person_id")["condition_start_datetime"]
    .min()
    .reset_index()
)

# Step 2: Create the event column (1 for all cases since all had cancer)
y_cases["event"] = 1  

y_cases
```

```python
# Define the columns to keep
columns_to_keep = [
    "person_id", 
    "date_of_birth"
]

# Filter the dataset
cancer_birthdate = dataset_13911935_person_df[columns_to_keep]

# Display the first few rows
cancer_birthdate.head(10)
```

```python
# get cancer outcome data with two dfs 
import pandas as pd
y_cases = pd.merge(y_cases, cancer_birthdate, on = "person_id")
y_cases.head()
```

```python
# Convert to datetime (if not already)
y_cases["cancer_diagnosis_date"] = pd.to_datetime(y_cases["condition_start_datetime"])
y_cases["date_of_birth"] = pd.to_datetime(y_cases["date_of_birth"])

# Check for missing values
print(y_cases[["cancer_diagnosis_date", "date_of_birth"]].isnull().sum())

# Ensure correct calculation of time_to_event
y_cases["time_to_event"] = (y_cases["cancer_diagnosis_date"] - y_cases["date_of_birth"]).dt.days / 365.25

# Check min and max values
print(y_cases["time_to_event"].describe())
```

```python
# Count the number of cases where time_to_event is suspiciously low
low_cases_count = (y_cases["time_to_event"] < 20).sum()

print(f"Number of suspiciously low cases: {low_cases_count}")
```

```python
def consolidate_gender(gender, sex_at_birth):
    # Convert inputs to lowercase strings and strip spaces
    gender = str(gender).lower().strip()
    sex_at_birth = str(sex_at_birth).lower().strip()

    # Initialize an empty set to store categories
    categories = set()

    # Define gender categories explicitly
    gender_map = {
        "gender identity: female": "female",
        "female": "female",
        "gender identity: male": "male",
        "male": "male",
        "gender identity: non binary": "nonbinary",
        "nonbinary": "nonbinary",
        "gender identity: additional options": "nonbinary",
        "gender identity: transgender": "transgender",
        "transgender": "transgender"
    }

    # Define sex at birth categories explicitly
    sex_at_birth_map = {
        "sex at birth: female": "female",
        "female": "female",
        "sex at birth: male": "male",
        "male": "male",
        "intersex": "intersex",
        "sex at birth: sex at birth none of these": "unknown"
    }

    # Define unknown category phrases
    unknown_terms = {"pmi: skip", "i prefer not to answer", "not man only, not woman only, prefer not to answer, or skipped"}

    # Assign categories from gender column
    if gender in gender_map:
        categories.add(gender_map[gender])
    elif gender in unknown_terms:
        categories.add("unknown")

    # Assign categories from sex at birth column
    if sex_at_birth in sex_at_birth_map:
        categories.add(sex_at_birth_map[sex_at_birth])
    elif sex_at_birth in unknown_terms:
        categories.add("unknown")

    # If both are unknown, return ["unknown"]
    if categories == {"unknown"}:
        return ["unknown"]

    # If one is unknown, discard it to keep the known value(s)
    categories.discard("unknown")

    return list(categories)  # Return as a list
```

```python
def consolidate_race_ethnicity(race, ethnicity):
    race = str(race).lower().strip()
    ethnicity = str(ethnicity).lower().strip()

    # Set to store categories
    categories = set()

    # Race categories (check for substrings, not exact matches)
    if "white" in race:
        categories.add("white")
    if "black" in race or "african american" in race:
        categories.add("black")
    if "asian" in race:
        categories.add("asian")
    if "american indian" in race or "alaska native" in race:
        categories.add("american_indian_alaskan_native")
    if "middle eastern" in race or "north african" in race:
        categories.add("middle_eastern_north_african")
    if "hawaiian" in race or "pacific islander" in race:
        categories.add("hawaiian_pacific_islander")
    if "more than one" in race:
        categories.add("multiple_races")

    # Ethnicity categories (Fixing the issue)
    if ethnicity == "hispanic or latino":  
        categories.add("hispanic_latino")

    # If both race and ethnicity are missing, mark as "unknown"
    if not categories:
        return ["unknown_race"]

    return list(categories)  # Return as a list to allow multiple labels

```
